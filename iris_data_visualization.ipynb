## Data Visualization
Data visualization helps transform raw data into meaningful graphical representations. In this project, different charts are used to understand the distribution of data, compare species, and identify relationships between various flower measurements.

### Count Plot

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Species',
    data=df
)

plt.title("Number of Flowers in Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.savefig("images/count_plot.png")
plt.show()
```

### Histogram
A histogram is used to visualize the distribution of numerical features such as sepal length. It helps understand the frequency and spread of the data.

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["SepalLengthCm"],
    bins=20,
    edgecolor="black"
)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.savefig("images/hist_plot.png")
plt.show()
```

### Scatterplot
A scatter plot is used to examine the relationship between two numerical features. Different colors represent different flower species, making it easier to identify clusters and patterns.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="SepalLengthCm",
    y="PetalLengthCm",
    hue="Species",
    data=df,
    s=80
)

plt.title("Sepal Length vs Petal Length")
plt.savefig("images/scatter_plot.png")
plt.show()
```

### Box Plot
A box plot summarizes the distribution of numerical data using quartiles. It helps identify the median, spread of data, and potential outliers for each species.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Species",
    y="PetalLengthCm",
    data=df
)

plt.title("Petal Length Across Species")
plt.savefig("images/box_plot.png")
plt.show()
```

### Pair Plot
A pair plot visualizes relationships between all numerical features simultaneously. It is useful for comparing feature distributions and understanding how different species are separated.

```python
sns.pairplot(
    df.drop("Id", axis=1),
    hue="Species"
)
plt.savefig("images/pair_plot.png")
plt.show()
```

### Heat Map
A correlation heatmap displays the correlation coefficients between numerical features. It helps identify strong positive or negative relationships among the variables.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.drop(["Id", "Species"], axis=1).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.savefig("images/heat_map.png")
plt.show()
```
